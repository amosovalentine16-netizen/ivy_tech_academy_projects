```python
##### To store student information create a dictionary
students={}
#enter the number of students
number_students=int(input("Enter the number of students:"))
#introduce for loops to record marks
for values in range(number_students):
    name=input("\Enter student name:")
    marks=[]
#for recorded subjects maths,chem,bio,english append marks
    for subject in ["Math", "Chem", "Bio","English"]:
        mark=float(input(f"Enter{subject}mark:"))
        marks.append(mark)
#calculte average
average=sum(marks)/len(marks)
#Assign letter grades using if,elif and else(control flow)
if average>=90:
    grade="A"
elif average>=80:
    grade="B"
elif average>=70:
    grade="C"
elif average>=60:
    grade="D"
elif average>=50:
    grade="E"
else:
    grade="F"
#store the students data(nest a dictionary for each student) Average stores average score,Marks stores marks list and Grade stores the grade
students[name]={"Marks":marks,"Average":average,"Grade":grade}
#loop through the students dictionary to print each students report.
for name,info in students.items():
    print(f"\nstudent:{name}")
    print(f"Marks:{info["Marks"]}")
    print(f"Average:{info["Average"]:.2f}")#2f prints average with 2dp
    print(f"Grade:{info["Grade"]}")
```
